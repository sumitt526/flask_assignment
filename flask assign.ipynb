flask is a lightweight web application framework designed to make getting started quick and easy with ability to scale upto complex applications. flask is classified as a micro framework, it has it's own set of tools,technologies and libraries to help with web application developement.

some advantages of using flask framework are:->

1) lightweight and modular:-> flask allows developers to use only the components they need, resulting in smaller and more maintainable web applications.
2) easy to learn:-> flask's simplicity and minimalistic approach make it an excellent choice for beginners.
3) flexible:-> flask can be easily extended with third party libraries and packages to accomodate a wide range of requirements.
4) simplicity of developement:-> flask is built around the core concepts of simplicity and flexibility and has extensive documentation allowing even newer developers to build web apps and APIs quickly.
5) control:-> Flask framework enables a programmer to do a lot more and make a lot more decisions in the custom way he or she desires. This reduces a number of problems that arise due to the rigidity of project layout in other frameworks.

App routing in Flask is used to map specific urls to associated functions that handle the logic for those urls. this technique helps users remember application urls and makes it easier to access specific pages directly without having to navigate from the home page.

We use app routes to define how our flask application should respond to different url requests. For example, we can define a route for the URL /hello that displays a “Hello, World!” message when a user visits that url. To bind a function to a url path we use the app.route decorator.

```python
# Qsn 5 solution:->

#In flask the url_for() function is used for building urls dynamically.It accepts the name of a function as its first argument and any number
#of keyword arguments,each corresponding to a variable part of the url.
#Here’s an example that demonstrates how the url_for() function can be used in a Flask application:->

from flask import Flask, url_for
app = Flask(__name__)

@app.route('/')
def index():
    return 'Index Page'

@app.route('/hello')
def hello():
    return 'Hello World!'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('hello'))
    
#In this example we have two routes defined: / and /hello, which are associated with the index and hello functions respectively. We use the
#url_for() function to build urls for these routes by passing the names of their associated functions as arguments. When we run this code it
#will print the URLs for the index and hello routes.
```
